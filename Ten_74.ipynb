```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_df = pd.read_csv(url, names=column_names)
```

```python
# 1. List down the features and their types
print("Features and their types:")
print(iris_df.dtypes)
```

```python
# 2. Create a histogram for each feature
iris_df.hist(figsize=(10,8))
plt.suptitle("Histograms of Iris Dataset Features")
plt.show()
```

```python
# 3. Create a boxplot for each feature
plt.figure(figsize=(10,8))
sns.boxplot(data=iris_df)
plt.title("Boxplot of Iris Dataset Features")
plt.show()
```

```python
# 4. Identify outliers
plt.figure(figsize=(12,6))
sns.boxplot(data=iris_df.drop(columns=['class']))
plt.title("Boxplot without 'class' feature to identify outliers")
plt.show()
```
